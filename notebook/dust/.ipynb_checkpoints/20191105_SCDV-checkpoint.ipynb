```python
matrix = tokenizer.texts_to_matrix(words, "count") # 出現回数
```

```python
# sns.heatmap(pd.DataFrame(sequence))
sns.heatmap(pd.DataFrame(matrix))
```

```python
import gensim
word_vectors = gensim.models.KeyedVectors.load_word2vec_format('model/model_fasttext_neo.vec', binary = False)
```

```python
WORD_DIM = len(tokenizer.word_index)
#配布されていたfastTextの次元数は300だったので300
EMBEDDING_DIM = 300
embedding_metrix = np.zeros((WORD_DIM+1,EMBEDDING_DIM))
#embedding_metrix = np.zeros((WORD_DIM, EMBEDDING_DIM))
```

```python
tokenizer.word_index.items()
```

```python
hoge = 0
piyo = 0
for word, i in tokenizer.word_index.items():
    try:
        hoge += 1
        embedding_vector = word_vectors[word]
        embedding_metrix[i-1] = embedding_vector
    except KeyError:
        piyo += 1
        # word_vectorsにないwordについては正規分布に従う乱数をダミーとして挿入
        embedding_metrix[i-1] = np.random.normal(0, np.sqrt(0.25), EMBEDDING_DIM)
print(hoge, piyo)
```

```python
sns.heatmap(pd.DataFrame(embedding_metrix))
```

```python
# np.dot(sequence, embedding_metrix)
np.dot(matrix, embedding_metrix)
```

```python
X_w2v = np.dot(matrix, embedding_metrix) # 文章ベクトル
```

```python
sns.heatmap(pd.DataFrame(X_w2v))
```

行列の積を求めることは、文章中の各単語のベクトル（日本語wiliによるfasttextの事前学習ベクトル）を足し合わせていることと同義

### word2vec + SCDV

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(words)
words_counts = tokenizer.texts_to_matrix(words, "count") # 文書中の出現回数
import gensim
word_vectors = gensim.models.KeyedVectors.load_word2vec_format('model/model_fasttext_neo.vec', binary = False)
word_kinds = len(tokenizer.word_index) # 単語ベクトルデータ中の単語種類数
word_vec_dim = 300 # 単語ベクトルの次元数
```

```python
# データサイズを一時的に小さくする
# words = words[:10]
```

```python
ex_words = set(word_vectors.index2word)
used_words = set([w for doc in words for w in doc])
```

```python
ex_used_words = list(ex_words & used_words)
used_word_vectors = np.array([word_vectors[w] for w in ex_used_words])
from sklearn import mixture
clusters_num = 30 # 実際はもっと大きくすべき
gmm = mixture.GaussianMixture(n_components=clusters_num, covariance_type='tied', max_iter=50)
gmm.fit(used_word_vectors)
word_probs = gmm.predict_proba(used_word_vectors)
```

```python
def idf(word, document):
    n_samples = len(document)
    df = np.sum(np.array([int(word in d) for d in document], dtype="float32"))
    n_samples += 1.0
    df += 1.0
    return np.log(n_samples / df) + 1.0
```

```python
word_cluster_vectors = np.zeros(shape=(len(ex_used_words), clusters_num*word_vec_dim))
for i, used_word in enumerate(ex_used_words):
  if i%100==0:
    print(i, end=' ')
  tmp = np.array([])
  for j in range(clusters_num):
    tmp = np.hstack([tmp, idf(used_word, words)*word_probs[i][j]*word_vectors[used_word]])
  word_cluster_vectors[i] = tmp
```

```python
doc_vectors = np.zeros(shape=(len(words), clusters_num*word_vec_dim))
for i, doc in enumerate(words):
  print(i, end=' ')
  for w in doc:
    if not w in ex_used_words:
      continue
    doc_vectors[i] += word_cluster_vectors[ex_used_words.index(w)]
for i in range(len(words)):
  doc_vectors[i] = doc_vectors[i]/np.linalg.norm(doc_vectors[i])
```

```python
pd.DataFrame(doc_vectors).to_csv(f"midway/{infile_name}_words_scdv.csv", index=0, header=None)
```

```python
# 必要なライブラリを読み込み
!pip install lightgbm
import lightgbm as lgb
```

```python
def lgbm_modeling(model_name):
  """学習、重要度可視化、モデルdump"""
  # 分類器の呼び出し 
  lgbm = lgb.LGBMClassifier(random_state=0, n_estimators=100)

  # 学習
  lgbm.fit(X_train, y_train)
  
  print(lgbm)

  # 重要度の抽出
  feature_importances = pd.Series(lgbm.feature_importances_).sort_values(ascending=True)
  feature_importances = feature_importances.to_frame(name='重要度').sort_values(by='重要度', ascending=False)

  word2id = word_to_id.join(feature_importances)

  feature_importances = feature_importances.merge(word_to_id, how='inner', left_index=True, right_on=['id'])
  feature_importances = feature_importances.set_index('word')

  feature_importances = feature_importances.drop("id", axis=1)

  # TOP20可視化
  feature_importances[0:20].sort_values(by='重要度').plot.barh()
  plt.legend(loc='lower right')
  plt.title('重要度TOP20')

  path = f"model/{model_name}"
  if not os.path.exists(path):
    os.makedirs(path)
  
  plt.savefig(f"{path}/{model_name}_重要度.png",
                bbox_inches="tight", dpi=150)
  plt.show()

  pickle.dump(lgbm, open(f"{path}/{model_name}.pkl", 'wb'))

  feature_importances.to_csv(f"{path}/{model_name}_重要度.csv", encoding='cp932')
```

```python
from sklearn.metrics import classification_report

def model_pred(model_name):
  """モデルload、予測、精度確認"""
  
  # モデルload
  path = f"model/{model_name}"
  model = pickle.load(open(f"{path}/{model_name}.pkl", 'rb')) 
  print(model)
  print('---'*35)
  
  # 予測結果確認
  pred = model.predict(X_test)
  print('予測結果：', pred[0:5])
  print('---'*35)
  
  prob = model.predict_proba(X_test)
  print('予測確率：', prob[0:5])
  print('---'*35)
  
  # 精度確認
  target_names = list(set(labels))

  print(y_test)

  dict_cr = classification_report(np.hstack(y_test), np.hstack(pred), 
                                target_names=target_names,
                                output_dict=True)

  # dfで表示
  df_cr = pd.DataFrame(dict_cr)

  display(
    df_cr[target_names].round(3),
    df_cr[['accuracy']].reset_index(drop=True)[0:1].round(3)
  )
  
  return pred, prob, df_cr
```

```python
from sklearn.model_selection import train_test_split

def make_train_test(vector_type):
  words = pd.read_csv(f"midway/{infile_name}_words_{vector_type}.csv", header=None)

  X_train, X_test, y_train, y_test = train_test_split(
      words, labels, test_size=0.25, random_state=0, stratify=labels)
  return X_train, X_test, y_train, y_test
```
